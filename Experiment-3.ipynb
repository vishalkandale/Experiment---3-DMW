```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
```

```python
basket = [['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
          ['Eggs','Milk','Onion','Apple','Ice cream','Eggs'],
          ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
          ['Milk','Apple','Kidney Beans','Eggs'],
          ['Ice cream','Onion','Kidney Beans','Eggs'],
          ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'],
          ['Onion','Onion','Apple','Ice cream','Yogurt'],
          ['Yogurt','Eggs','Onion','Kidney Beans','Ice cream','Eggs'],
          ['Kidney Beans','Onion','Yogurt','Kidney Beans','Eggs'],
          ['Corn','Onion','Onion','Kidney Beans','Ice cream','Eggs']
         ]
```

```python
basket
```

```python
TE = TransactionEncoder()
TE_ary =TE.fit(basket).transform(basket)
```

```python
df = pd.DataFrame(TE_ary,columns=TE.columns_)
```

```python
df.head()
```

```python
frequent_itemsets = apriori(df,min_support=0.6,use_colnames=True)
frequent_itemsets.head()
```
